```python
import numpy as np
import matplotlib.pyplot as plt
from matplotlib.gridspec import GridSpec
import os
```

```python
# PARAMETERS

save_to_disk = False # If set to True, figure will be saved to disk. Set to false if you don't want figure to be saved to disk.
```

```python
#Loading Data

# Load numerical simulation data points: provided to us courtesy of Jacob Shen from the S2023 paper.
data_c_10 = np.genfromtxt(os.path.join('..', 'Data', 'S2023 Fig 3, c=10.dat'), dtype=None, delimiter='') 
data_c_30 = np.genfromtxt(os.path.join('..', 'Data', 'S2023 Fig 3, c=30.dat'), dtype=None, delimiter='')
data_c_100 = np.genfromtxt(os.path.join('..', 'Data', 'S2023 Fig 3, c=100.dat'), dtype=None, delimiter='')

# Load survival fractions computed using the K2021 method
old_method_curve_c_10 = np.load(r"survival_fractions_K2021_c_10_ErrorChecking.npy")
old_method_curve_c_30 = np.load(r"survival_fractions_K2021_c_30_ErrorChecking.npy")
old_method_curve_c_100 = np.load(r"survival_fractions_K2021_c_100_ErrorChecking.npy")

# Load the survival fractions using our switching procedure
switching_methods_curve_c_10 = np.load(r"survival_fractions_switching_c_10_ErrorChecking.npy")
switching_methods_curve_c_30 = np.load(r"survival_fractions_switching_c_30_ErrorChecking.npy")
switching_methods_curve_c_100 = np.load(r"survival_fractions_switching_c_100_ErrorChecking.npy")

# Load the survival fractions computed using S2023 fitting functions
S2022_curve_c_10 = np.load(r"survival_fractions_S2023_c_10_ErrorChecking.npy")
S2022_curve_c_30 = np.load(r"survival_fractions_S2023_c_30_ErrorChecking.npy")
S2022_curve_c_100 = np.load(r"survival_fractions_S2023_c_100_ErrorChecking.npy")
```

```python
#Computing absolute error between K2021 method and S2023 numerically simulated data
old_method_errors_c_10 = old_method_curve_c_10 - data_c_10[:,1]
old_method_errors_c_30 = old_method_curve_c_30 - data_c_30[:,1]
old_method_errors_c_100 = old_method_curve_c_100 - data_c_100[:,1]

#Computing absolute error between our switching procedure and S2023 numerically simulated data
switching_methods_errors_c_10 = switching_methods_curve_c_10 - data_c_10[:,1]
switching_methods_errors_c_30 = switching_methods_curve_c_30 - data_c_30[:,1]
switching_methods_errors_c_100 = switching_methods_curve_c_100 - data_c_100[:,1]

#Computing absolute error between S2023 fitting functions and S2023 numerically simulated data
S2022_errors_c_10 = S2022_curve_c_10 - data_c_10[:,1]
S2022_errors_c_30 = S2022_curve_c_30 - data_c_30[:,1]
S2022_errors_c_100 = S2022_curve_c_100 - data_c_100[:,1]



#Computing Relative error between K2021 method and S2023 numerically simulated data
old_method_rel_errors_c_10 = old_method_errors_c_10 / data_c_10[:,1]
old_method_rel_errors_c_30 = old_method_errors_c_30 / data_c_30[:,1]
old_method_rel_errors_c_100 = old_method_errors_c_100 / data_c_100[:,1]

#Computing Relative error between our switching procedure and S2023 numerically simulated data
switching_methods_rel_errors_c_10 = switching_methods_errors_c_10 / data_c_10[:,1]
switching_methods_rel_errors_c_30 = switching_methods_errors_c_30 / data_c_30[:,1]
switching_methods_rel_errors_c_100 = switching_methods_errors_c_100 / data_c_100[:,1]

#Computing absolute error between S2023 fitting functions and S2023 numerically simulated data
S2022_rel_errors_c_10 = S2022_errors_c_10 / data_c_10[:,1]
S2022_rel_errors_c_30 = S2022_errors_c_30 / data_c_30[:,1]
S2022_rel_errors_c_100 = S2022_errors_c_100 / data_c_100[:,1]
```

```python
plt.rcParams["mathtext.fontset"] = "cm"

fig = plt.figure()
gs = GridSpec(2, 2, height_ratios=[1, 4])

ax = fig.add_subplot(gs.new_subplotspec((0, 0), colspan=2))
ax2 = fig.add_subplot(gs.new_subplotspec((1, 0), colspan=2))

# plot the same data on both axes

ax.plot(data_c_100[:,0], np.zeros(np.size(data_c_100[:,0])), color="grey", linestyle="dashed")
ax2.plot(data_c_100[:,0], np.zeros(np.size(data_c_100[:,0])), color="grey", linestyle="dashed")

ax.scatter(data_c_10[:,0], old_method_rel_errors_c_10, 80*np.ones(np.size(data_c_10[:,0])), marker="o", color="green", facecolors='none')
ax.scatter(data_c_30[:,0], old_method_rel_errors_c_30, 80*np.ones(np.size(data_c_30[:,0])), marker="o", color="red", facecolors='none')
ax.scatter(data_c_100[:,0], old_method_rel_errors_c_100, 80*np.ones(np.size(data_c_100[:,0])), marker="o", color="blue", facecolors='none')
ax2.scatter(data_c_10[:,0], old_method_rel_errors_c_10, 80*np.ones(np.size(data_c_10[:,0])), marker="o", color="green", facecolors='none')
ax2.scatter(data_c_30[:,0], old_method_rel_errors_c_30, 80*np.ones(np.size(data_c_30[:,0])), marker="o", color="red", facecolors='none')
ax2.scatter(data_c_100[:,0], old_method_rel_errors_c_100, 80*np.ones(np.size(data_c_100[:,0])), marker="o", color="blue", facecolors='none')

ax.scatter(data_c_10[:,0], switching_methods_rel_errors_c_10, 80*np.ones(np.size(data_c_10[:,0])), marker="v", color="green", facecolors='none')
ax.scatter(data_c_30[:,0], switching_methods_rel_errors_c_30, 80*np.ones(np.size(data_c_30[:,0])), marker="v", color="red", facecolors='none')
ax.scatter(data_c_100[:,0], switching_methods_rel_errors_c_100, 80*np.ones(np.size(data_c_100[:,0])), marker="v", color="blue", facecolors='none')
ax2.scatter(data_c_10[:,0], switching_methods_rel_errors_c_10, 80*np.ones(np.size(data_c_10[:,0])), marker="v", color="green", facecolors='none')
ax2.scatter(data_c_30[:,0], switching_methods_rel_errors_c_30, 80*np.ones(np.size(data_c_30[:,0])), marker="v", color="red", facecolors='none')
ax2.scatter(data_c_100[:,0], switching_methods_rel_errors_c_100, 80*np.ones(np.size(data_c_100[:,0])), marker="v", color="blue", facecolors='none')

ax.scatter(data_c_10[:,0], S2022_rel_errors_c_10, 80*np.ones(np.size(data_c_10[:,0])), marker="^", color="green", facecolors='none')
ax.scatter(data_c_30[:,0], S2022_rel_errors_c_30, 80*np.ones(np.size(data_c_30[:,0])), marker="^", color="red", facecolors='none')
ax.scatter(data_c_100[:,0], S2022_rel_errors_c_100, 80*np.ones(np.size(data_c_100[:,0])), marker="^", color="blue", facecolors='none')
ax2.scatter(data_c_10[:,0], S2022_rel_errors_c_10, 80*np.ones(np.size(data_c_10[:,0])), marker="^", color="green", facecolors='none')
ax2.scatter(data_c_30[:,0], S2022_rel_errors_c_30, 80*np.ones(np.size(data_c_30[:,0])), marker="^", color="red", facecolors='none')
ax2.scatter(data_c_100[:,0], S2022_rel_errors_c_100, 80*np.ones(np.size(data_c_100[:,0])), marker="^", color="blue", facecolors='none')

ax.plot([], [], color="green", label=r"$c=10$")
ax.plot([], [], color="red", label=r"$c=30$")
ax.plot([], [], color="blue", label=r"$c=100$")
ax.scatter([], [], 80*np.ones(1), color="black", facecolors='none', label=r"K2021")
ax.scatter([], [], 80*np.ones(1), color="black", marker="v", facecolors='none', label=r"Switching procedure")
ax.scatter([], [], 80*np.ones(1), color="black", marker="^", facecolors='none', label=r"S2023 fitting function")

ax.set_xscale("log")
ax2.set_xscale("log")

ax2.set_xlabel(r"$E_{\rm frac}$", fontsize=18)
ax2.set_ylabel(r"Relative Error in SF", fontsize=18)

# zoom-in / limit the view to different portions of the data
ax.set_ylim(6.45, 6.7)  # outliers only
ax2.set_ylim(-0.3, 1.6)  # most of the data

ax.spines["bottom"].set_visible(False)
ax2.spines["top"].set_visible(False)

ax.tick_params(axis="x", which="both", length=0)
ax.set(xticklabels=[])
ax.tick_params(labeltop="off")
ax2.tick_params(labeltop="off")
ax2.xaxis.tick_bottom()
ax.xaxis.tick_top()
ax.tick_params(labelright=True, right=True)
ax2.tick_params(labelright=True, right=True)



d = 0.25  # how big to make the diagonal lines in axes coordinates
# arguments to pass to plot, just so we don't keep repeating them
kwargs = dict(marker=[(-1, -d), (1, d)], markersize=12,
              linestyle="none", color='k', mec='k', mew=1, clip_on=False)
ax.plot([0, 1], [0, 0], transform=ax.transAxes, **kwargs)
ax2.plot([0, 1], [1, 1], transform=ax2.transAxes, **kwargs)

fig.legend(bbox_to_anchor=(0.37,0.65), loc="center", fontsize = 12)
fig.tight_layout()
if save_to_disk:
    fig.savefig("Figure (8).pdf")
plt.show()
```
