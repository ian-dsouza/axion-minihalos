```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
# PARAMETERS
save_to_disk = False # If set to True, figure will be saved to disk. Set to false if you don't want figure to be saved to disk.
```

```python
E_fracs = np.load("Efracs_switching.npy") # values of E_frac used to generate the single encounter case

survival_fractions_full_curve = np.load(r"survival_fractions_switching_c_100.npy") # Values of survival fractions for the single encounter case with concentration = 100
survival_fractions_Efrac_7Point8_multiple_encounters_fixed_Efrac_switching = np.load(r"survival_fractions_Efrac_7Point8_multiple_encounters_fixed_Efrac_switching.npy") # this is the multiple encounter case

E_fracs_incremental_7Point8 = [7.8, 7.8*2, 7.8*3, 7.8*4, 7.8*5] # Cumulative E_frac values for the multiple encounter case where incremental E_frac = 7.8

survival_fractions_for_effective_single_encounter = np.load(r"survival fractions for effective single encounter; incremental E_frac = 7.8.npy") # Values of survival fractions using effective single encounter scenario with 
                                                                                                                                                # effective value of 'p'
```

```python
plt.rcParams["mathtext.fontset"] = "cm"

plt.loglog(E_fracs, survival_fractions_full_curve, label=r"Analytical NFW ($c=100$)")
plt.loglog(E_fracs_incremental_7Point8, survival_fractions_Efrac_7Point8_multiple_encounters_fixed_Efrac_switching, color="red", marker="o", label=r"Multiple encounter with $E_{\rm frac} = 7.8$")

num_points=10


plt.loglog(E_fracs_incremental_7Point8, survival_fractions_for_effective_single_encounter, color="orange", marker="^", label=r"Effective single encounter with $E_{\rm frac} = 7.8$")

plt.tick_params(which="both", labelright=True, right=True)
plt.xlabel(r"$E_{\rm frac}$", fontsize=18)
plt.ylabel(r"$1 - \frac{\Delta M}{M_{\rm vir,s}}$", fontsize=18)
# plt.title(r"Survival fraction vs. $E_{\mathrm{frac}}$ (multiple encounters)", fontsize=18)
plt.xlim(10**(-2), 10**2)
plt.legend(fontsize = 11)
plt.tight_layout()
if save_to_disk:
    plt.savefig("Figure 11; bottom panel.pdf")
plt.show()
```
