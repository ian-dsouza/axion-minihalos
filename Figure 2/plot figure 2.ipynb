```python
import numpy as np
import matplotlib.pyplot as plt
import math
```

```python
plt.rcParams["mathtext.fontset"] = "cm"

c_val = 10
f_NFW_val = math.log(1+c_val)-c_val/(1+c_val)
alpha_squared = 3/c_val**2 + 1/(2*f_NFW_val)*(c_val-3)/(c_val+1)
beta = (c_val**3 - 2*c_val*(1+c_val)*f_NFW_val) / (2*(1+c_val)**2 * f_NFW_val**2)

x = np.linspace(0, 1, 500)
psi = 1 / f_NFW_val * np.log(1+c_val*x) / x

Es_frac = np.array([0.1, 1, 10])
linestyles = ["dotted", "dashdot", "dashed"]
labels = [r"$\vert\Delta\epsilon(x)\vert$; $E_{\rm frac}=0.1$", r"$\vert\Delta\epsilon(x)\vert$; $E_{\rm frac}=1$", r"$\vert\Delta\epsilon(x)\vert$; $E_{\rm frac}=10$"]

plt.plot(x, psi, color="green", label=r"$\psi(x)$")

for i, E_frac in enumerate(Es_frac):
    
    delta_eps = beta / alpha_squared * E_frac * x**2

    plt.plot(x, delta_eps, color="green", linestyle=linestyles[i], label=labels[i])

plt.xlabel(r"$x$", fontsize=18)
plt.xlim(0, 1)
plt.ylim(0, 15)
plt.legend(fontsize = 15)
plt.tight_layout()
#plt.savefig("psi crossing over deltaEpsilon for three values of E_frac and c = 10.pdf")
plt.show() 
```
